```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import io
```

```python
#reading the uploaded file
df=pd.read_csv(r"C:\\Users\\himanshu rai\\Desktop\\node\\finalProject\\Crime rate detection data.csv")
```

```python
df.insert(19,"SafeMeasure",0)
```

```python
for i in range(0,len(df)):
    if df.Crime_rate[i]>=150:
      df.SafeMeasure[i]=0
    elif df['Crime_rate'][i]>100 and df['Crime_rate'][i]<150:
      df.SafeMeasure[i]=1
    else:
      df.SafeMeasure[i]=2
```

```python
x = df[['GDPP','TotalPop','MALE_LIT']]
y= df['SafeMeasure']
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2,test_size=0.2,shuffle=True)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```

```python
#for kanpur(should be moderate safe)
cap=87296*0.36
pop=3234000*0.87
lit=88.34*0.96
gdpp=cap/pop
result=clf.predict([[gdpp,pop,lit]])
if result==0:
    print("not safe")
elif result==1:
  print("moderate safe")
else:
  print("safe")
```
