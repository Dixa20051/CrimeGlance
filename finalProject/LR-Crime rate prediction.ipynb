```python
x = df[['GDPP','TotalPop','MALE_LIT']]
y= df['Crime_rate']
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2,test_size=0.2,shuffle=True)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.predict(x_test)
```

```python
reg.score(x_test,y_test)
```

```python
#for kanpur(should be moderate safe)
cap=87296*0.36
pop=3234000*0.87
lit=88.34*0.96
gdpp=cap/pop
result=reg.predict([[gdpp,pop,lit]])
if result>150:
  print("not safe")
elif result>100 and result<150:
  print("moderate safe")
else:
  print("safe")
```
